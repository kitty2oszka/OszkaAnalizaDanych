```python
data = eurostat.get_data_df('DEMO_GIND')
data.info()
```

```python
data.head(n=10)
```

```python
data = data.rename(columns={r'geo\TIME_PERIOD': 'geo'})

df_long = data.melt(id_vars=["freq", "indic_de", "geo"], var_name='Rok', value_name='value')
df_long = df_long[df_long['value'].notna() & (df_long['value'] != ':')]


df_filtered = df_long[(df_long['geo'] == 'EA19') & (df_long['indic_de'] == 'AVG')]
fig = px.line(df_filtered, x='Rok', y='value')
fig.show()
```

Zad 2.2

```python
print(df_long['indic_de'].unique())
```

```python
kraje = ['PL', 'DE', 'CZ', 'FR', 'SE']
```

```python
df_pol = df_long[(df_long['indic_de']=='DEATH') & (df_long['geo'].isin(kraje))]

df_pol.loc[:,'value'] = df_pol['value'].astype(float)
df_pol.loc[:,'Rok'] = df_pol['Rok'].astype(int)



fig = px.line(df_pol, x='Rok', y='value', color = 'geo',title="Porównanie liczby zgonów: Polska vs DE, CZ, FR, SE",labels={"geo": "Kraj", "Value": "Liczba zgonów", "Rok": "Rok"})
fig.show()
```

```python
df_long['Rok'] = df_long['Rok'].astype(int)
```

```python
df_ludi = df_long[(df_long['indic_de'] == 'POPT') & (df_long['Rok'] == 2023)]
df_ludi_sorted = df_ludi.sort_values('value', ascending=False)
df_ludi_sorted.head(n=20)
```

```python
fig4 = px.pie(df_ludi_sorted, values='value', names='geo', title='Top 20 państw z największą liczbą ludności w 2023 roku')
fig4.update_layout(margin=dict(t=200))
fig4.update_layout(width=700, height=700)

fig4.show()
```

zad 2.4

```python
df_pop_2023 = df_long[(df_long['indic_de'] == 'POPT') & (df_long['Rok'] == 2023)].copy()
pop_pol = df_pop_2023.loc[df_pop_2023['geo'] == 'PL', 'value'].values[0]

df_pop_2023.loc[:, 'diff_to_pol'] = (df_pop_2023['value'] - pop_pol).abs()

closest_countries = df_pop_2023.sort_values('diff_to_pol').head(4)
print(closest_countries[['geo', 'value', 'diff_to_pol']])
```

Zad 2.5

```python
df_birth_death = df_long[df_long['indic_de'].isin(['LBIRTH', 'DEATH'])]
df_eu_parts = df_birth_death[((df_birth_death['geo'] == 'EU27_2007') & (df_birth_death['Rok'] >= 2007) & (df_birth_death['Rok'] <= 2013)) |((df_birth_death['geo'] == 'EU28') & (df_birth_death['Rok'] > 2013) & (df_birth_death['Rok'] <= 2020)) |
((df_birth_death['geo'] == 'EU27_2020') & (df_birth_death['Rok'] > 2020) & (df_birth_death['Rok'] <= 2023))]
```

```python
df_eu_parts.loc[:, 'geo'] = 'EU'
```

```python
df_eu_combined = df_eu_parts.groupby(['geo', 'indic_de', 'Rok'], as_index=False)['value'].sum()
fig5 = px.line(df_eu_combined, x='Rok', y='value', color='indic_de', title = 'Liczba narodzin i zgonów w UE (2007-2023)')
fig5.show()
```
